### Section 1

This section pulls in the Public Census Data Status_DSTP2024 Excel spreadsheet completed by the 2024 DSTP Cohort and creates a data frame of just the sources needed for Team Discover's domains (General; Infrastructure, Housing, and Transportation; and Health and Nutrition). Columns are based off the Metadata Template Excel file.

```python
#Read-in Public Census Data Status Excel file
df=pd.read_excel('Public Census Data Status_DSTP2024.xlsx')

#Get column headers
df_headers=df.iloc[0]
#print(df_headers)

#Add column headers to metadata
all_sources=df[3:]
all_sources.columns=df_headers
#print(all_sources.head)

#Keep necessary columns
all_sources_trim=all_sources.drop(['data.census.gov','API','Neither','Team','Participant','Notes (add as many as you would like!)','How was the measure calculated','How do you interpret the statistic presented?'],axis=1)
#print(all_sources_trim.columns)

#Use column names designated by Cass and Emily
column_names=['Source','Data_Description','Data_Provenance','Data_Link','Metadata_Link','Coverage','Frequency','Unit','Geography','Uncertainty','Citation']

#Keep necessary rows
td_sources=all_sources_trim[all_sources_trim['Surveys and Programs (https://www.census.gov/programs-surveys/surveys-programs.html)'].str.contains('|'.join(sources))]
td_sources.columns=column_names
print(td_sources)
```

### Section 2

This section creates a data frame of Team Discover's measures. Columns are based off the Metadata Template Excel file.

```python
#Getting JSON file for 2022 ACS DP Table Variables from Census API
apiRequestObject=requests.get('https://api.census.gov/data/2022/acs/acs5/profile/variables.json')
apiRequestJsonObject=apiRequestObject.json()

#Turning JSON object into a data frame
metadata_all=pd.DataFrame(apiRequestJsonObject)

#Restricting data frame to selected measures
metadata_acsdp=metadata_all[metadata_all.index.str.contains('|'.join(measures))]
print(metadata_acsdp[0:2])
```

```python
#Getting JSON file for 2022 ACS Subject Table Variables from Census API
apiRequestObject=requests.get('https://api.census.gov/data/2022/acs/acs5/subject/variables.json')
apiRequestJsonObject=apiRequestObject.json()

#Turning JSON object into a data frame
metadata_all=pd.DataFrame(apiRequestJsonObject)

#Restricting data frame to selected measures
metadata_acssub=metadata_all[metadata_all.index.str.contains('|'.join(measures))]
print(metadata_acssub[0:2])
```

```python
#Getting JSON file for 2020 Decennial DHC Table Variables from Census API
apiRequestObject=requests.get('https://api.census.gov/data/2020/dec/dhc/variables.json')
apiRequestJsonObject=apiRequestObject.json()

#Turning JSON object into a data frame
metadata_all=pd.DataFrame(apiRequestJsonObject)

#Restricting data frame to selected measures
metadata_dec=metadata_all[metadata_all.index.str.contains('|'.join(measures))]
print(metadata_dec.head)
```

```python
#Getting JSON file for 2021 CBP Table Variables from Census API
apiRequestObject=requests.get('https://api.census.gov/data/2021/cbp/variables.html')
html_string=apiRequestObject.text
#print(html_string)
import html_to_json
apiRequestJsonObject=html_to_json.convert(html_string)

#Turning JSON object into a data frame
metadata_all=pd.DataFrame(apiRequestJsonObject)

#Restricting data frame to selected measures
metadata_cbp=metadata_all[metadata_all.index.str.contains('|'.join(measures))]
print(metadata_cbp.head)
```

```python
#Create variables data frame
variables=pd.DataFrame(measures)
#print(variables[0:2])

#Create Variable_Names column
variables['Variable_Name']=pd.DataFrame(measurenames)

#Create Variable_Labels column based on Variable_Names column
variables['Variable_Label']=variables['Variable_Name'].str.replace("_"," ")\
    .str.replace('non ','non-')\
    .str.replace('non-Hispanic','non-Hispanic,')\
    .str.replace('AIAN','American Indian and Alaska Native')\
    .str.replace('NHPI','Native Hawaiian and Pacific Islander')\
    .str.replace('SOR','Some Other Race')\
    .str.replace('Multi alone','Two or More Races')\
    .str.replace('Percent no ','Percent With No ')\
    .str.replace('Number employed ','Number employed in ')\
    .str.replace('Percent p','Percent With P')\
    .str.replace('Percent With Public t','Percent Public T')\
    .str.replace('oil kerosene etc','Oil, Kerosene, Etc.')\
    .str.title()\
    .str.replace('Lp','LP')\
    .str.replace('With','with')\
    .str.replace('And','and')\
    .str.replace('Or','or')\
    .str.replace('Of','of')\
    .str.replace('In ','in ')\
    .str.replace(' To',' to')

#Create blank columns for Variable_Descriptions and Interpretation
variables['Variable_Description']=variables[0]
variables['Interpretation']=""

#Create Domain column based on row numbers
variables['Domain']='General'
variables['Domain'][nmeasurenames_gen:]='Infrastructure, Housing, & Transportation'
variables['Domain'][nmeasurenames_iht:]='Health & Nutrition'

#Create Source column
variables['Source Abbrev']='ACS'

#Update column names
variables.columns=['API Variable','Variable_Name','Variable_Label','Variable_Description','Interpretation','Domain','Source Abbrev']

#Print preview
print(variables)
```
